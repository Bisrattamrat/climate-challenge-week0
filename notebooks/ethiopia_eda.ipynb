```python
df = pd.read_csv("../data/ethiopia.csv")
df.head()
```

```python
df["Date"]=pd.to_datetime(df["YEAR"]*1000+df["DOY"],format="%Y%j")
df["Month"]= df["Date"].dt.month
```

```python
df.replace(-999,np.nan,inplace=True)
```

```python
df= df.drop_duplicates()
```

```python
df.describe()
```

```python
df.isna().sum()
(df.isna().sum()/len(df))*100
```

The dataset contains no missing values across all columns(0% missing)
This indicates that the dataset is complete and does not require imputation or removal of null values at this stage. The absence of missing data improves the reliability of the analysis and ensures that no bias is introduced due to data gaps

```python
from scipy import stats

cols =["T2M","T2M_MAX","T2M_MIN","PRECTOTCORR","RH2M","WS2M","WS2M_MAX"]

z_scores= np.abs(stats.zscore(df[cols]))
outliers=(z_scores>3)
outliers.sum()
```

A total of 137 outliers were detected across the selected climate virables using the Z-score method(|Z|>3)
These outliers may represent extreme weather conditions such as unusually high tempratures, heavy rainfall, or strong winds,which are important in climate analysis 
Therefor, the outliers willbe retained rather than removed, as they may carry meaningful information about extreme climate events.

```python
df.ffill(inplace=True)
```

```python
df.to_csv("../data/ethiopia_clean.csv", index=False)
```

```python
df_monthly= df.groupby("Month")["T2M"].mean()
df_monthly.plot(title="Monthly Avg Temperature(Ethiopia)")
```

No extreme anomalies are immediately visible, indicating relatively stable temperature trends across the observed years.

```python
df_rain =df.groupby("Month")["PRECTOTCORR"].sum()
df_rain.plot(kind="bar",title="Monthly Rainfall (Ethiopia)")
```

The monthly rainfall pattern in Ethiopia shows a clear seasonal trend.Rainfall gradually increases from the beginning of the year and peaks around July-August, indicating the main rainy season.After this peak, rainfall declines sharply towards the end of the year, with the lowest levels observed between November and February .This suggests a distinct wet and dry season pattern.

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

The correlations heatmap shows relatioships between climate variables.Strong postive Correlations are observed between T2M,T2M_MAX and T2M-MIN,indicating that tempreature variables move together.There is also a noticeable relationship between temperature and humidity, While wind speed virables show moderate correlation.These relationships help identify how different climate factors interact.

```python
sns.scatterplot(x="T2M",y="RH2M",data=df )
sns.scatterplot(x="T2M_RANGE",y="WS2M",data=df)
```

The scatter plot shows a weak relationship between temperature(T2M) and humidity(RH2M), with no strong linear pattern observed.

```python
sns.scatterplot(x="T2M_RANGE",y="WS2M", data=df)
```

The scatter plot shows a moderate positive relationship between temperature range(T2M_RANGE) and wind speed(WS2M),indicating that higher temperature variation is associated with stronger winds.
